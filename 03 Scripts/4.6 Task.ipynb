# 03. Merge and Check for Full Match

- Note common column headers between two dfs from head() code results above
- Use 'product_id' for merge key
- Use an Inner Join to ensure dataframe is complete without missing information - will include only information that has data from both frames in the merge

```python
# Test merge

pd.merge(df_ords_prods, df_prods, on = 'product_id')
```

```python
# Merge df_ords_prods with df_prods using inner join merge and merge flag

df_ords_prods_merged = df_ords_prods.merge(df_prods, on = 'product_id', indicator = True)
```

```python
df_ords_prods_merged.head()
```

```python
# Check the merge flag to ensure all rows from both dfs were included

df_ords_prods_merged['_merge'].value_counts()
```

**Confirmed** Merge only includes data that was in both dfs - count of rows is different than those of the individual dfs
